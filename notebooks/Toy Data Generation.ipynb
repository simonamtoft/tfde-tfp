```python
# Number of points generated from each dataset
N = 3000
```

```python
# Import and plot data 

f, ax = plt.subplots(2, 5, figsize=(20, 10))

# checkerboard 
data = gen_checkerboard(batch_size=N)
x = data[:, 0]
y = data[:, 1]
ax[0, 0].plot(x, y, '.')
ax[0, 0].set_title('Checkerboard')

# 8 gaussians
data_8 = gen_8gaussians(batch_size=N)
x_8 = data_8[:, 0]
y_8 = data_8[:, 1]
ax[0, 1].plot(x_8, y_8, '.')
ax[0, 1].set_title('8gaussians')

# pinwheel
data = gen_pinwheel(batch_size=N)
x = data[:, 0]
y = data[:, 1]
ax[1, 0].plot(x, y, '.')
ax[1, 0].set_title('pinwheel')

# moons
data = gen_moons(batch_size=N)
x = data[:, 0]
y = data[:, 1]
ax[1, 1].plot(x, y, '.')
ax[1, 1].set_title('moons')

# swissroll
data = gen_swissroll(batch_size=N)
x = data[:, 0]
y = data[:, 1]
ax[0, 2].plot(x, y, '.')
ax[0, 2].set_title('swissroll')

# circles
data = gen_circles(batch_size=N)
x = data[:, 0]
y = data[:, 1]
ax[0, 3].plot(x, y, '.')
ax[0, 3].set_title('circles')

# rings
data = gen_rings(batch_size=N)
x = data[:, 0]
y = data[:, 1]
ax[0, 4].plot(x, y, '.')
ax[0, 4].set_title('rings')

# 2spirals
data = gen_2spirals(batch_size=N)
x = data[:, 0]
y = data[:, 1]
ax[1, 2].plot(x, y, '.')
ax[1, 2].set_title('2spirals')

# line
data = gen_line(batch_size=N)
x = data[:, 0]
y = data[:, 1]
ax[1, 3].plot(x, y, '.')
ax[1, 3].set_title('line')

# cos 
data = gen_cos(batch_size=N)
x = data[:, 0]
y = data[:, 1]
ax[1, 4].plot(x, y, '.')
ax[1, 4].set_title('cos')

plt.show()
```
